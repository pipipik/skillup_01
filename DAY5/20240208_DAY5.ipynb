```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(4, 4))
plt.plot([0,2,4,6,8],[0,6,12,24,48], label="graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("title")
plt.legend()
plt.show
```

```python
x = np.linspace(0, 10, 20)
y = x**2
z = (x - 1)**3
w = ((x - 2)*(x - 4))**2
plt.plot(x, y, label="x**2")
plt.plot(x, z, label="(x-1)**2")
plt.plot(x, w, label="((x-2)*(x-4))**2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show
```

```python
plt.figure(figsize = (12,4))
plt.subplot(1,3,1)
plt.plot(x,y,label="y=x^2")
plt.legend()

plt.subplot(1,3,2)
plt.plot(x,z,label="y=(x-1)^3")
plt.legend()

plt.subplot(1,3,3)
plt.plot(x,w,label="y=((x-2)(x-4))^2")
plt.legend()

plt.tight_layout()
```
