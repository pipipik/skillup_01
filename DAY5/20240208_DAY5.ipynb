```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(4, 4))
plt.plot([0,2,4,6,8],[0,6,12,24,48], label="graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("title")
plt.legend()
plt.show
```

```python
x = np.linspace(0, 10, 20)
y = x**2
z = (x - 1)**3
w = ((x - 2)*(x - 4))**2
plt.plot(x, y, label="x**2")
plt.plot(x, z, label="(x-1)**2")
plt.plot(x, w, label="((x-2)*(x-4))**2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show
```

```python
plt.figure(figsize = (12,4))
plt.subplot(1,3,1)
plt.plot(x,y,label="y=x^2")
plt.legend()

plt.subplot(1,3,2)
plt.plot(x,z,label="y=(x-1)^3")
plt.legend()

plt.subplot(1,3,3)
plt.plot(x,w,label="y=((x-2)(x-4))^2")
plt.legend()

plt.tight_layout()
```

```python
# 3x3でサイズ指定された空のfigureを作成
fig=plt.figure(figsize=(3,3))

# figの中に、1行1列のaxesをfigの中に作成
ax=fig.add_subplot(1,1,1)

# axの中に実際にグラフを描画
ax.plot(x,y,label="y=x^2")
ax.plot(x,z,label="y=(x-1)^3")
ax.plot(x,w,label="y=((x-2)(x-4))^2")

# xの描画範囲を0, 10に設定
# axesに対して設定する場合はset_が先頭につくことに注意
ax.set_xlim(0,10)
ax.legend()
```

```python
# 1行3列のaxesを生成
fig,ax=plt.subplots(1,3,figsize=(14,4))
print(ax)#axはarray形式であることを確認
ax[0].plot(x,y,label="y=x^2")
ax[1].plot(x,z,label="y=(x-1)^3")
ax[2].plot(x,w,label="y=((x-2)(x-4))^2")
# for分で三つのグラフを順番に描画
for i in range(3):
    ax[i].legend()
    ax[i].set_xlabel("x axis")
    ax[i].set_ylabel("y axis")
plt.tight_layout()
```

```python
# 2行2列のaxesを生成
fig,ax=plt.subplots(2,2)

# np.array形式であることを利用して描画
ax[0,0].plot(x,y,label="y=x^2")
ax[0,0].legend()
ax[0,1].plot(x,z,label="y=(x-1)^3")
ax[0,1].legend()
ax[1,0].plot(x,w,label="y=((x-2)(x-4))^2")
ax[1,0].legend()
```

```python
# Matplotlibの基本事項
# 問題1
x = np.linspace(-np.pi, np.pi, 20)
y_sin = np.sin(x)
y_cos = np.sin(x)
y_tan = np.tan(x)
fig, ax = plt.subplots(3, 1, figsize=(8, 6))
ax[0].plot(x, y_sin, label = "y=sin(x)", color = "red")
ax[1].plot(x, y_cos, label = "y=cos(x)", color = "blue")
ax[2].plot(x, y_tan, label = "y=tan(x)", color  = "green")

for i in range(3):
    ax[i].legend()
plt.tight_layout()
```
